```python
import pandas as pd
import numpy as np
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
```

```python
df = pd.read_csv("Sample Apple vs. Orange.csv")
```

```python
df
```

```python
processed_df = pd.DataFrame()
```

```python
for label, _ in df.items():
    if (label == "Thickness of Skin (mm)"):
        continue
    processed_df[label] = LE.fit_transform(df[label])
processed_df['Thickness of Skin (mm)'] = df['Thickness of Skin (mm)']
```

```python
processed_df = processed_df.sample(frac=1)
processed_df
```

```python
X = processed_df.drop("Fruit Type", axis=1)
y = processed_df['Fruit Type']
# 1 = orange
```

```python
c_coef = 50
```

```python
clf = SVC(gamma='auto', kernel='linear', C=c_coef)
```

```python
clf.fit(X, y)
```

```python
clf.coef_
```

```python
clf_without_hair = SVC(gamma='auto', kernel='linear', C=c_coef)
clf_without_hair.fit(X.drop("Owner's Name", axis=1), y)
```

```python
clf_without_hair.coef_
```

```python
with_hair = np.around(clf.coef_, decimals=3)
without_hair = np.around(clf_without_hair.coef_, decimals=3)
```

```python
print("With Hair Variable | Without Hair Variable")
print(f"Weight of Skin Bumpiness:\t{with_hair[0][0]} | {without_hair[0][0]}")
print(f"Weight of Location Grown:\t{with_hair[0][1]} | {without_hair[0][1]}")
print(f"Weight of Owner's Name:\t\t{with_hair[0][2]}")
print(f"Weight of Thickness of Skin:\t{with_hair[0][3]} | {without_hair[0][2]}")
```
